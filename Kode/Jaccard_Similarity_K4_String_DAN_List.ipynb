```python
import pandas as pd
```

```python
df = pd.read_csv('/content/BPOM_Tokped_Blocked_K4.csv')
df.head(10)
```

```python
set('bless acne face powder')
```

```python
df['ltable_Nama_Produk'] = df['ltable_Nama_Produk'].astype('str')
df['rtable_Nama Produk'] = df['rtable_Nama Produk'].astype('str')
df.dtypes
```

```python
def jaccard_similarity(a, b):
    # convert to set
    a = set(a)
    b = set(b)
    # calucate jaccard similarity
    j = float(len(a.intersection(b))) / len(a.union(b))
    return j
```

```python
jaccard_score = []
for index, row in df.iterrows():
  jscore = jaccard_similarity(row['ltable_Nama_Produk'], row['rtable_Nama Produk'])
  jaccard_score.append(jscore)

df['Jaccard_Score'] = jaccard_score
df.head(10)
```
