```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Tokped_Face_Powder.csv')
df.head()
```

```python
df['Nama Produk'] = df['Nama Produk'].str.replace(r'[^\w\s]+', '')
df.head(100)
```

```python
df['Nama Produk'] = df['Nama Produk'].str.lower()
df.head()
```

```python
df['Merk'] = df['Nama Produk'].str.split(' ').str[0]
df.head()
```

```python
from google.colab import files

df.to_csv('Tokped_Face_Powder_Preprocessed.csv', encoding = 'utf-8-sig')
files.download('Tokped_Face_Powder_Preprocessed.csv')
```
