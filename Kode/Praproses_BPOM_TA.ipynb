```python
import pandas as pd
```

```python
df = pd.read_csv('/content/bpom_serbuktabur.csv')
df.head()
```

```python
#remove non-word character and whitespace
df['Nama_Produk'] = df['Nama_Produk'].str.replace(r'[^\w\s]+', '')
df.head(100)
```

```python
df['Merk'] = df['Merk'].str.replace(r'[^\w\s]+', '')
df.head(100)
```

```python
#lowercase kolom Nama_Produk
df['Nama_Produk'] = df['Nama_Produk'].str.lower()
df.head()
```

```python
#lowercase kolom Nama_Produk
df['Merk'] = df['Merk'].str.lower()
df.head()
```

```python
from google.colab import files

df.to_csv('bpom_serbuktabur_preprocessed.csv', encoding = 'utf-8-sig')
files.download('bpom_serbuktabur_preprocessed.csv')
```
